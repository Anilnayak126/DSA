```python
def locate_cards(cards, query):
   
    
    lo, hi = 0, len(cards) -1

    while lo <= hi:
        mid = lo + (hi - lo) // 2
        mid_number = cards[mid]

        print("lo", lo , ", hi", hi, ",mid" , mid, ", mid number",mid_number)
        
        if mid_number == query:
            return mid
        elif mid_number  < query:

            hi = mid - 1 
        elif mid_number > query:
            
            lo = mid + 1
            
    return -1
```

####  Finding the First or Last Occurrence in a Sorted Array

Problem:

- Given a sorted array with duplicates, find the index of the first occurrence of a target value.

##### Explanation:


- Perform binary search as usual.

- When the target is found, instead of returning the index, continue searching in the left half to find the first occurrence and for right half  for last.



```python
def BinarySearch(arr,target):

    def first_occurance(arr,target):
        
        left,right = 0 , len(arr) -1

        first  = -1

        while left <= right:
            
            mid = (left + right) // 2

            if arr[mid] == target:
                
                first = mid

                right = mid - 1


            elif arr[mid] < target:

                left = mid + 1

            else:
                right = mid - 1

        return  first 
    
    def  find_last_occurance(arr,target):

        left,right = 0 , len(arr) -1

        last = -1 

        while left <=  right:

            mid = (left + right)  // 2 

            if arr[mid] == target:
                last = mid

                left = mid + 1

            elif arr[mid] < target:
                left = mid + 1
            else: 
                right = mid - 1

        return last
    first = first_occurance(arr,target)
    last = find_last_occurance(arr,target)

    return f"firt {first} ------ last {last}"


arr = [1,2,3,4,5,5,5,5,6,8]

print(BinarySearch(arr,5))
```

#### Finding the Peak Element in an Array
A peak element is an element that is greater than or equal to its neighbors.

##### Problem:

- Given an array of integers, find a peak element. Assume that the array may contain multiple peaks, and return any one of them.

##### Explanation:
- Use binary search to find the peak.

- Compare the middle element with its neighbors.

- If the middle element is smaller than the right neighbor, a peak must exist in the right half.

- Otherwise, a peak must exist in the left half.
