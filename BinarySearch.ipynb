## Binary Search

```python
def locate_cards(cards, query):
   
    
    lo, hi = 0, len(cards) -1

    while lo <= hi:
        mid = (lo + hi) // 2
        mid_number = cards[mid]

        print("lo", lo , ", hi", hi, ",mid" , mid, ", mid number",mid_number)
        
        if mid_number == query:
            return mid
        elif mid_number  < query:

            hi = mid - 1 
        elif mid_number > query:
            
            lo = mid + 1
            
    return -1
```

```python

tests = []
test = {
    'input' : {
        'cards':[12,34,45,23],
        'query' : 45
    },
    'output' :2
}

tests.append(test)


# quert occurs in the middle

tests.append(test)

tests.append({
    'input':{
        'cards':[12,34,1,2,3,9],
        'query':1
    },
    'output': 2
})


# query in the last element 

tests.append({
    'input':{
        'cards':[12,34,1,2,3,-9],
        'query':-9
    },
    'output': 5
})

#  query in the first element 
tests.append({
    'input':{
        'cards':[12,34,1,2,3,9],
        'query':12
    },
    'output': 0
})


#cards contains both query and cards 
tests.append({
    'input':{
        'cards':[2],
        'query' :2,
    },
    'output': 0
})

# if card  doesnt have any cards like this
tests.append({
    'input':{
        'cards':[12,34,2,45],
        'query':3,
    },
    'output': -1
})

tests.append({
    'input':{
        'cards':[],
        'query':3,
    },
    'output': -1
})

# number can repeat in cards

tests.append({
    'input':{
        'cards':[12,23,12,34,12,12],
        'query':34,
    },
    'output': 3
})


# query occurs multiple time

tests.append({
    'input':{
        'cards':[12,23,12,34,12,12],
        'query':12,
    },
    'output': 0
})

```

```python
for test in tests:
    print(locate_cards(test['input']['cards'],test['input']['query']) == test['output'])
```

```python
from jovian.pythondsa import evaluate_test_cases

evaluate_test_cases(locate_cards,tests)
```
