### Circular Linked List 

```python
class Node: 
    def __init__(self,data):
        self.data = data
        self.next = None
    

class CircularLinkedLIst:

    def __init__(self):
        self.head = None
        self.tail = None

    

    def insert_at_beginning(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return 
        new_node.next = self.head 
        self.head = new_node
        self.tail.next = self.head


    def insert_at_end(self,data):
        new_node = Node(data)
        if self.head is None: 
            self.head = new_node
            self.tail = new_node 
            return
        temp = self.head
        while temp:
            temp = temp.next
            if temp == self.head: 
                break 
        self.tail.next = new_node
        self.tail =new_node
        self.tail.next = self.head
    


    def insert_at_pos(self,data,pos):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return
        temp = self.head

        for i in range(pos - 1):
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

    def delete_at_beginning(self):
        if self.head is None:
            print('CLL is empty..')
            return
        temp = self.head
        self.head = temp.next 
        temp.next = None
        self.tail.next = self.head


    def delete_at_end(self):
        if  self.head  is None: 
            print('CLL is empty..')
            return
        temp = self.head.next
        prev = self.head

        while temp:
            temp = temp.next
            prev = prev.next
            if self.head == temp.next: 
                break
        prev.next = None
        self.tail = prev
        self.tail.next = self.head

    def delete_at_the_pos(self,pos):
        if self.head is None: 
            print("CLL is Empty..")
            return
        temp  =self.head.next
        prev = self.head

        for i in range(pos - 1):
            temp = temp.next
            prev = prev.next
        prev.next = temp.next
        temp.next = None
        
    def display(self):
        if self.head is None: 
            print("CLL is empty..")
            return
        temp = self.head

        while temp:
            print(temp.data,end='-->')
            temp = temp.next
            if temp == self.head:
                break
        print(temp.data)


cll = CircularLinkedLIst()

cll.insert_at_beginning(23)
cll.insert_at_beginning(26)
cll.insert_at_beginning(45)
cll.insert_at_end(12)
cll.insert_at_pos(34,2)
# cll.delete_at_beginning()
# cll.delete_at_end()
cll.delete_at_the_pos(2)


cll.display()
```
